Q3. xplain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

TRIM: Removes extra spaces from text.
 >>Ex: " Hello " → "Hello"

LEFT: Takes a certain number of characters from the start of a text.
>>Ex: LEFT("India123", 5) → "India"

RIGHT: Takes characters from the end of a text.
>>Ex: RIGHT("India123", 3) → "123"

MID: Extracts text from the middle of a string.
>>Ex: MID("ABC123XYZ", 4, 3) → "123"

CONCAT: Joins two or more pieces of text together.
>>Ex: CONCAT("Hello", " ", "World") → "Hello World"

Q8. Understanding TODAY() and NOW()

a. What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use
 each function.

>> TODAY() - It gives only the current date.Time not included.

   Ex: To calculate age or number of days left for an event.



NOW() - It gives current date AND time.

   Ex: To record the exact time a report was created or updated.


 b. If cell A1 contains the date 2025-06-10, write a formula using TODAY() to determine how many days are left until that date
